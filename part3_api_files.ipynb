```python
# ==============================
# TASK 1 — FILE READ & WRITE
# ==============================

# ------------------------------
# Part A — Write to file
# ------------------------------
try:
    # Write mode
    with open("python_notes.txt", "w", encoding="utf-8") as f:
        f.write("Topic 1: Variables store data. Python is dynamically typed.\n")
        f.write("Topic 2: Lists are ordered and mutable.\n")
        f.write("Topic 3: Dictionaries store key-value pairs.\n")
        f.write("Topic 4: Loops automate repetitive tasks.\n")
        f.write("Topic 5: Exception handling prevents crashes.\n")
    
    print("File written successfully.")

except Exception as e:
    print("Error while writing file:", e)


# Append mode
try:
    with open("python_notes.txt", "a", encoding="utf-8") as f:
        f.write("Topic 6: Functions help organize code.\n")
        f.write("Topic 7: APIs allow communication between systems.\n")
    
    print("Lines appended.")

except Exception as e:
    print("Error while appending:", e)


# ------------------------------
# Part B — Read file
# ------------------------------
try:
    with open("python_notes.txt", "r", encoding="utf-8") as f:
        lines = f.readlines()

    print("\n--- File Content ---")

    # Print numbered lines
    for i, line in enumerate(lines, start=1):
        print(f"{i}. {line.strip()}")

    # Count lines
    print(f"\nTotal lines: {len(lines)}")

    # Keyword search
    keyword = input("\nEnter keyword to search: ").lower()

    matches = [line.strip() for line in lines if keyword in line.lower()]

    if matches:
        print("\nMatches found:")
        for m in matches:
            print(m)
    else:
        print("No matching lines found.")

except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print("Error while reading file:", e)
```

```python
# ==============================
# TASK 2 — API INTEGRATION
# ==============================

import requests

BASE_URL = "https://dummyjson.com/products"

# ------------------------------
# Step 1 — Fetch and Display Products
# ------------------------------
try:
    response = requests.get(f"{BASE_URL}?limit=20")
    data = response.json()

    products = data["products"]

    print("\n--- Product List ---")
    print("ID  | Title                          | Category      | Price    | Rating")
    print("----|--------------------------------|---------------|----------|-------")

    for p in products:
        print(f"{p['id']:<4}| {p['title'][:30]:<30}| {p['category']:<13}| ${p['price']:<8}| {p['rating']}")

except Exception as e:
    print("Error fetching products:", e)


# ------------------------------
# Step 2 — Filter and Sort
# ------------------------------
try:
    filtered = [p for p in products if p["rating"] >= 4.5]

    sorted_products = sorted(filtered, key=lambda x: x["price"], reverse=True)

    print("\n--- Filtered (Rating ≥ 4.5) & Sorted by Price ---")

    for p in sorted_products:
        print(f"{p['id']} | {p['title']} | ${p['price']} | Rating: {p['rating']}")

except Exception as e:
    print("Error filtering/sorting:", e)


# ------------------------------
# Step 3 — Search by Category (Laptops)
# ------------------------------
try:
    response = requests.get(f"{BASE_URL}/category/laptops")
    data = response.json()

    print("\n--- Laptops Category ---")

    for p in data["products"]:
        print(f"{p['title']} - ${p['price']}")

except Exception as e:
    print("Error fetching laptops:", e)


# ------------------------------
# Step 4 — POST Request (Simulated)
# ------------------------------
try:
    payload = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    response = requests.post(f"{BASE_URL}/add", json=payload)

    print("\n--- POST Response ---")
    print(response.json())

except Exception as e:
    print("Error in POST request:", e)
```

```python
# ==============================
# TASK 3 — EXCEPTION HANDLING
# ==============================

import requests

BASE_URL = "https://dummyjson.com/products"

# ------------------------------
# Part A — Guarded Calculator
# ------------------------------
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


print("\n--- Safe Divide Tests ---")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# ------------------------------
# Part B — Guarded File Reader
# ------------------------------
def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")


print("\n--- File Reader Tests ---")
print(read_file_safe("python_notes.txt"))   # should work
print(read_file_safe("ghost_file.txt"))     # should fail gracefully


# ------------------------------
# Part C — Robust API Calls
# ------------------------------
def fetch_products():
    try:
        response = requests.get(f"{BASE_URL}?limit=20", timeout=5)
        response.raise_for_status()
        return response.json()["products"]

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


def fetch_laptops():
    try:
        response = requests.get(f"{BASE_URL}/category/laptops", timeout=5)
        response.raise_for_status()
        return response.json()["products"]

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


def create_product():
    try:
        payload = {
            "title": "My Custom Product",
            "price": 999,
            "category": "electronics",
            "description": "A product I created via API"
        }

        response = requests.post(f"{BASE_URL}/add", json=payload, timeout=5)
        response.raise_for_status()

        print("\n--- POST Response ---")
        print(response.json())

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


# Run API functions safely
products = fetch_products()

if products:
    print("\n--- Products Fetched ---")
    for p in products[:5]:  # show first 5 only
        print(p['title'])

laptops = fetch_laptops()
if laptops:
    print("\n--- Laptops ---")
    for l in laptops:
        print(l['title'], "-", l['price'])

create_product()


# ------------------------------
# Part D — Input Validation Loop
# ------------------------------
while True:
    user_input = input("\nEnter a product ID (1–100) or 'quit' to exit: ")

    if user_input.lower() == "quit":
        print("Exiting program.")
        break

    # Validate number
    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    if not (1 <= product_id <= 100):
        print("Invalid ID. Must be between 1 and 100.")
        continue

    try:
        response = requests.get(f"{BASE_URL}/{product_id}", timeout=5)

        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            data = response.json()
            print(f"{data['title']} - ${data['price']}")
        else:
            print("Unexpected response:", response.status_code)

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)
```

```python
# ==============================
# TASK 4 — ERROR LOGGING SYSTEM
# ==============================

import requests
from datetime import datetime

BASE_URL = "https://dummyjson.com/products"

# ------------------------------
# Logger Function
# ------------------------------
def log_error(context, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {context}: {error_type} — {message}\n"

    with open("error_log.txt", "a", encoding="utf-8") as f:
        f.write(log_entry)


# ------------------------------
# 1. Trigger ConnectionError
# ------------------------------
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    print("Connection error triggered.")
    log_error("fetch_products", "ConnectionError", str(e))
except Exception as e:
    log_error("fetch_products", "Exception", str(e))


# ------------------------------
# 2. Trigger HTTP Error (404)
# ------------------------------
try:
    response = requests.get(f"{BASE_URL}/999", timeout=5)

    if response.status_code != 200:
        print("HTTP error triggered (404).")
        log_error(
            "lookup_product",
            "HTTPError",
            f"{response.status_code} Not Found for product ID 999"
        )

except requests.exceptions.ConnectionError as e:
    log_error("lookup_product", "ConnectionError", str(e))
except requests.exceptions.Timeout as e:
    log_error("lookup_product", "Timeout", str(e))
except Exception as e:
    log_error("lookup_product", "Exception", str(e))


# ------------------------------
# 3. Display Log File Contents
# ------------------------------
print("\n--- Error Log Contents ---")

try:
    with open("error_log.txt", "r", encoding="utf-8") as f:
        print(f.read())
except FileNotFoundError:
    print("No log file found.")
```
